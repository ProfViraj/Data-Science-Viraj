```python
for i in Titanic_training.isna().sum():
    print((i/len(Titanic_training)*100))
# 0 to 5% missing values : drop rows ['Embarked']
# 6 to 45 % missing values : replace the missing values ['Age']
# 50 % and more missing values : drop the column ['Cabin']
```

```python
# Dropping the column which has more than 50% missing values
Titanic_training.drop(columns='Cabin', inplace=True)
Titanic_training.isna().sum()
# All the missing data is treated
```

```python
X_train = Encoded_titanic_train.iloc[:,1:]  # Independent variable set
Y_train = Encoded_titanic_train.iloc[:,0]   # Dependent variable set
```

# Interview Questions:
1. What is the difference between precision and recall?

Ans. Precision is a conditional probability that the actual value is positive given that the predicted value is positive, on the other hand, recall is a conditional probability that the predicted value is positive given that the actual value is positive.

Calculation of recall contains both negative and positive values from CM, but in calculation on precision, only positive values are considered from CM.

2. What is cross-validation, and why is it important in binary classification?

Ans. Cross validation is a technique used to evaluate the performance of a model on testing data. It divide the available data into multiple folds or subsets, using one of these folds as a validation set, and training the model on the remaining folds. This process is repeated multiple times, each time using a furthur fold as the validation set. At last, the results from each validation step are meaned or combined to produce a more accurate model performance measure.
