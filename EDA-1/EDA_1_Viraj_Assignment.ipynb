d) Outliers detection and treatment

```python
sns.boxplot(df)
```

```python
def extreme_detection(data, colname):
    q1 = data[colname].quantile(0.25)
    q3 = data[colname].quantile(0.75)
    iqr = q3 - q1

    upper_extreme = q3 + (1.5 * iqr)
    lower_extreme = q1 - (1.5 * iqr)

    return lower_extreme, upper_extreme
```

```python
for i in df.columns:
  print(i+':',extreme_detection(df,i))
```

```python
for i in df.columns:
  lower_extreme = extreme_detection(df,i)[0]
  upper_extreme = extreme_detection(df,i)[1]
  df.loc[df[i]<lower_extreme, i] = lower_extreme
  df.loc[df[i]>upper_extreme, i] = upper_extreme
```

```python
sns.boxplot(df)
# Outliers treated
```

#	Statistical Summary:
-	Provide a statistical summary for each variable in the dataset, including measures of central tendency (mean, median) and dispersion (standard deviation, interquartile range).
-	Highlight any interesting findings from this summary.


```python
stats_summary = df.describe()
IQR={}
for i in df.columns:
  IQR[i]= df[i].quantile(0.75) - df[i].quantile(0.25)
IQR_df = pd.DataFrame(IQR, index=['IQR'])
stats_summary = pd.concat([stats_summary, IQR_df])
stats_summary
```

On above statistics, some interesting findings can be highlighted:

1. Variability in LB (FHR Baseline): The range of values for the FHR Baseline (LB) is relatively narrow, with a minimum of 105 and a maximum of 161. However, the interquartile range (IQR) is relatively wide at 14, indicating significant variability in this measure.

2. Low Variation in FM, UC, DL: The standard deviations for the features FM (FHR Modulations), UC (Uterine Contractions), and DL (Light Decelerations) are very low, indicating low variability in these measures across the dataset.

3. Negative Minimum for Several Features: Some features, such as ALTV (Average Long Term Variability) and MLTV (Mean Long Term Variability), have negative minimum values, which might indicate data anomalies or errors in measurement. Further investigation into these negative values is warranted.

4. Strong Positive Skewness in Several Features: Features like LB, ASTV (Percentage of Time with Abnormal Short Term Variability), and Width exhibit strong positive skewness, indicating that their distributions are skewed towards higher values. This skewness might suggest underlying patterns or characteristics in the data that merit closer examination.

5. Uniform Distribution of DS and DP: The features DS (Severe Decelerations) and DP (Prolongued Decelerations) have a standard deviation of 0, indicating that all values in the dataset are identical for these features. This uniform distribution suggests that these features may not provide much discriminatory information for analysis.

6. Consistent Median for NSP: The feature NSP (Fetal State Class Code) has a median value of 1 across the dataset, indicating that the majority of observations fall into this class. This consistent median suggests a relatively balanced distribution of fetal state classes in the dataset.
